```python
# ANSI color codes
RED = "\033[91m"
YELLOW = "\033[93m"
BLUE = "\033[94m"
GREEN = "\033[92m"
RESET = "\033[0m"

for i in range(13):
    for j in range(9):
        if (i==0 and (j==1 or j==2 or j==4 or j==6 or j==7)) or \
           (i==1 and (j==2 or j==3 or j==5 or j==6)) or \
           (i==2 and (j==3 or j==5)):
            print(" ", end=" ")

        elif (i==5 and (j==2 or j==6)) or \
             (i==6 and (j==1 or j==7)) or \
             (i==7 and (j==0 or j==8)) or \
             (i==10 and (j==3 or j==5)):
            print(" ", end=" ")

        elif (i==12 and (j in [1,2,3,4,5,6,7])) or \
             (i==11 and (j==2 or j==3 or j==5 or j==6)):
            print(" ", end=" ")

        else:
            # Color logic for butterfly parts
            if i < 4:
                print(RED + "*" + RESET, end=" ")
            elif 4 <= i <= 8:
                print(BLUE + "*" + RESET, end=" ")
            else:
                print(YELLOW + "*" + RESET, end=" ")

    print()
```
